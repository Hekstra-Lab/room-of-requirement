### Distribution Explorator
Just a couple small examples of using interact to look at different distributions as a function of their parameters. 

```python
%matplotlib notebook
from matplotlib import pyplot as plt
import numpy as np
from tensorflow_probability import distributions as tfd
from ipywidgets import interact
```

```python
xmin,xmax = 0., 10.
npoints = 10000

x = np.linspace(xmin, xmax, npoints)
dist = tfd.Gamma(5.,5.)
y = dist.prob(x)

f = plt.figure()
plt.title("Gamma Distribution")
line = plt.plot(x, y)[0]
ax = plt.gca()

@interact(conc=(0., 100.), rate=(0., 100.))
def update(conc, rate):
    dist = tfd.Gamma(conc, rate)
    y = dist.prob(x)
    line.set_data(x, y)

```

```python
xmin,xmax = 0., 10.
npoints = 10000

x = np.linspace(xmin, xmax, npoints)
dist = tfd.TruncatedNormal(5.,0.5, 0., 10.)
y = dist.prob(x)

f = plt.figure()
plt.title("Truncated Normal")
line = plt.plot(x, y)[0]
ax = plt.gca()

@interact(loc=(0., 10.), scale=(0., 2.))
def update(loc, scale):
    dist = tfd.TruncatedNormal(loc, scale, 0., 10.)
    y = dist.prob(x)
    line.set_data(x, y)

```
